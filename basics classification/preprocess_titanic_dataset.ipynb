```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic_data = pd.read_csv('archive/titanic_train.csv')
titanic_data.head()
```

```python
titanic_data.isna().sum()
```

```python
#we could impute a value for age, but for the sake of no complexty we will drop nan values
#we could also pick features based on correlation and we will also ignore that for now
titanic_data.drop(['Name','PassengerId','Cabin','Ticket'],inplace=True,axis=1)
titanic_data.dropna(inplace=True)
titanic_data.isna().sum()
```

```python
titanic_data.shape
```

```python
titanic_data.sample(5)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
titanic_data['Sex'] = labelencoder.fit_transform(titanic_data['Sex'])
titanic_data.head()
```

```python
labelencoder.classes_
```

```python
titanic_data = pd.get_dummies(titanic_data,columns=['Embarked'])
titanic_data.head()
```

```python
#randomise and drop for the old index
titanic_data = titanic_data.sample(frac=1).reset_index(drop=True)
titanic_data.head()
```

```python
titanic_data.to_csv('archive/titanic_data_proccess.csv',index=False)
```
